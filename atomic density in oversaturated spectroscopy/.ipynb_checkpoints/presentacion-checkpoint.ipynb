## 1. Introducción

Este reporte se adentrará en el estudio de la **espectroscopía saturada**, una técnica fundamental en la metrología cuántica y la física atómica, con un enfoque particular en el isótopo Rubidio-87 (Rb87). La espectroscopía saturada permite la resolución de transiciones hiperfinas en gases atómicos, eliminando el ensanchamiento Doppler que normalmente oscurecería estas características. La capacidad de caracterizar con precisión la densidad atómica y la temperatura de una muestra de Rb87 es crucial para una variedad de aplicaciones, desde la estabilización de frecuencias de láser hasta el desarrollo de relojes atómicos y la investigación en condensados de Bose-Einstein. A lo largo de este documento, exploraremos los principios detrás de esta técnica, cómo se utilizan los espectros de saturación para inferir las propiedades del vapor atómico y la importancia de estos parámetros en el contexto experimental.

## 3. Planteamiento del Problema

La caracterización precisa de las propiedades de un gas atómico, como su densidad atómica y temperatura, es un prerrequisito fundamental para la mayoría de los experimentos de física atómica y óptica cuántica. En particular, para el Rubidio-87, que es un átomo clave en campos como los **relojes atómicos, la metrología de frecuencia y la computación cuántica**, la habilidad de conocer estos parámetros es crítica. Sin embargo, la medición directa de la densidad y la temperatura en celdas de vapor es a menudo compleja debido a las condiciones de vacío y a la necesidad de no perturbar el sistema.

Tradicionalmente, la absorción lineal de la luz se ve afectada por el ensanchamiento Doppler, lo que hace que las líneas espectrales individuales de las transiciones hiperfinas del Rb87 se superpongan, impidiendo una resolución precisa y, por ende, una caracterización exacta de las propiedades del vapor. Esto lleva a la siguiente pregunta central:

*¿Cómo se puede utilizar la técnica de espectroscopía saturada para caracterizar de manera precisa la densidad atómica y la temperatura de un vapor de Rubidio-87, superando las limitaciones en la absorción lineal?*

## 4. Fundamentos Teóricos

### 4.1 Espectroscopía de Absorción

#### 4.1.1 Suceptibilidad eléctrica

De acuerdo Sierra [1], tenemos que la suceptibilidad eléctrica para un átomo de dos niveles como función de la desintonía está dada por,
$$ \chi(\Delta) = \frac{Nd_{21}^2}{\epsilon_0 \hbar} \frac{i}{\Gamma_2 - i \Delta} =  \frac{Nd_{21}^2}{\epsilon_0 \hbar} f(\Delta)$$
en donde $f(\Delta)$ es un factor de forma de lína de perfil Lorenziano (Sierra, 2018).

#### 4.1.2 Coeficiente de absorción 

Por la ley de Beer-Lambert sabemos que la transmisión $\mathscr{T}$ en términos de la desintonía está dado por,
$$\mathscr{T}(\Delta) = \frac{I_{out}}{I_{in}} = e^{-\alpha(\Delta)L}$$
donde el coeficiente de absorción $\alpha$ estpa definido como,
$$\alpha(\Delta) = klm(\chi(\Delta))$$

#### 4.1.3 Factores de intensidad relativa

Los valores  de los factores de intensidad relativa $C^2_F$ de las transiciones relevantes para el rubidio se presentan a continuación,
| Isótopo | Fg |  Fe  | $C^2_F$ |
| ------- | ---| ---- |---------|
| $^{87}$Rb | 2  | 1, 2 |  5/18, 5/18   |
| $^{87}$Rb | 1  | 1, 2 |  1/18, 5/18   |

#### 4.1.4 Ensanchamiento doppler

En cualquier vapor atómico a temperatura ambiente los átomos poseen velocidades de cientos de metros sobre segundo. Esto provocaque la frecuencia absorbida por los átomos sea despazada por $\Delta_D$ debido alefecto Doppler,
$$\Delta_D = - \vec{k}\cdot \vec{v} = -kv_z$$
donde $\vec{k}$ es el vector de onda del haz. Así, el factor de forma $f(\Delta)$ de la suceptibilidad se transforma en,
$$f(\Delta + \Delta_D) = \frac{i}{\Gamma_0/2 + i(\Delta - kv)}$$
tomando la parte imaginaria para la absorción,
$$f^I(\Delta +\Delta_D) = \frac{2}{\Gamma_0}[1+ (\frac{\Delta-kx}{\Gamma_0/2})^2]^{-1}$$
La velocidad térmica de los átomos está dada por la distribución de Maxwell-Boltzan.Esta distribución $g_u(v)$ es Gaussiana de ancho $u= \sqrt{ak_BT/M_I}$ en donde $T$ es la temperatura, $k_B$ es la constante de Boltzmann y $M_I$ es la masa atómica del isótopo.
$$g_u(v) = \frac{1}{\sqrt{\pi u^2}}exp[-(\frac{v}{u})^2]$$
Dado que no es posible lidiar con velocidades individuales, integramos y obtenemos que,
$$s^I(\Delta) = \int_{-\infty}^{\infty} f^I(\Delta - kv)g_u(v)dv$$
Éste factor de forma es una convolución de una Gaussiana de los perfiles de velocidades y una Lorenziana de la forma de línea homogénea. Ésta convolución es un perfil de voigt. Para facilitar la programación hacemos las siguientes sustituciones, $y = \frac{\Delta}{ku}$, $a = \frac{\Gamma_0}{ku}$ y queda,
$$s_I(\Delta) ) \frac{\sqrt{\pi}}{2ku} e^{\frac{1}{4}(a-i2y)^2}[Erfc(\frac{a}{2}-iy)+e^{i2ay}Erfc(\frac{a}{2} + iy)]$$
en donde $Erfc(z)$ es la función de error complementaria definida como,
$$Erfc(z) = \frac{2}{\sqrt{\pi}} \int_{z}^{\infty} e^{-t^2}dt$$

#### 4.1.2.5 Densidad atómica

La presión a la cual se encuentra el vapor se puede derivar de la ecuación de Causius-Clapeyron, que describe la curva de coexistencia en un diagrama $P-T$,
$$\frac{dP}{dT} = \frac{L}{T\Delta \nu}$$
donde $P$ es la presión, $T$ es la temperatura , $L$ es el calor latente de vaporización por partícula y $\Delta \nu = \nu_g - \nu_{s(l)}$ es el cambio en volúmen por molécula cuando éstas pasan de estado sólido (o líquido) a estado gaseoso. Considerando $\Delta \nu \approx \nu_g$, suponiendo que el vapor se comporta como un gas ideal $p \nu_g = k_b T$ y sustituyendo en la ecuación de Clausius-Clapeyron,
$$\frac{dp}{p} = \frac{L}{k_B T^2}dT$$
Expandiendo el calor latente como función de la temperatura ($ L = L_0 + L_1 T + L_2T^2 + ...$) y recuperando los primeros tres términos de la expansión podemos integrar de ambos lados,
$$\log_{10}(P) = A + \frac{B}{T} + CT + D \log_{10}(T)$$
A partir de ésta expresión y la ecuación de gas ideal obtenemos,
$$\log_{10}(N(T)) = A + \frac{B}{T} + CT + D \log_{10}(T) - \log_{10}(k_BT)$$
donde $A,B,C$ y $D$ son parámetros que se obtienen de forma experimental. A continuación se presentan dos conjuntos de parametros citados comúnmente en trabajos relacionados con alcalinos.



#### 4.1.6 Espectro de absorción teórico

El coeficiente de absorción de toda la línea $D1$ es la suma de los coeficientes de cada una de las ocho transiciones $Fg \to Fe$. El espectro de absorción se obtiene al aplicar la ley de Beer-Lambert usando los coeficientes de absorción,
$$\mathscr{T}(\Delta) = exp[-L(\sum_{F_g,F_e} \alpha_{F_g,F_e}(\Delta))]$$
donde el coeficiente de absorción es,
$$\alpha_{F_g,F_e}(\Delta) = klm(\chi_{F_g,F_e}(\Delta))= k C_F^2 d^2 N \frac{s_I(\Delta)}{g_I \hbar \epsilon_0} $$
y $g_I =2(2I+1)$ es la degeneración del estado base del isótopo (que es $8$ para $^{87}Rb$).

## 6. Análisis de Datos y Caracterización

El programa realizado para el procesamiento de las señales de espectroscopía y la caracterización de la temperatura y la densidad atómica está formado por una serie fases parael procesamient de los datos obtenidos.
1. **Pre-procesamiento de los datos:** Extracción de datos, eliminación de ruido, corrección de la línea base, normalización de señal.
2. **Cálculo de temperatura y densidad esperados:** A partir de las amplitudes de las señales de absorción, la longitud de la celda y la temperatura esperada.
3. **Ajuste de los espectros:** Ajuste de perfiles de Voigt sobre el espectro haciendo uso de paquetrías de ajuste funcional. 
4. **Almacenamiento de información:** Extracción de los parámetros de ajuste (temperatura, densidad atómica y corrimiento horizontal), reportes de ajuste, gráficas de espectro y residuales

## 8. Conclusiones

Este reporte ha detallado la aplicación de la espectroscopía saturada como una técnica poderosa para la caracterización de la densidad atómica y la temperatura en un vapor de Rubidio-87 (Rb87). Se ha demostrado cómo, al superar las limitaciones del ensanchamiento Doppler, es posible resolver las finas estructuras hiperfinas del Rb87, lo que a su vez permite la inferencia precisa de las propiedades del vapor. Los principios teóricos de la absorción saturada, la relación entre las características del espectro y los parámetros del vapor, así como el análisis de datos, fueron presentados.

La espectroscopía saturada es una herramienta indispensable para obtener espectros de alta resolución de las transiciones hiperfinas del Rb87, eliminando el ensanchamiento Doppler. El análisis de la amplitud de las señales de absorción saturada permite una determinación robusta de la densidad atómica del vapor de Rb87. La densidad atomica del vapor puede ser inferida con precisión a partir de la relación entre la temperatura y la presión de vapor saturado del Rubidio. La caracterización precisa de estos parámetros es fundamental para una amplia gama de aplicaciones en física atómica y metrología, asegurando la optimización y el éxito de experimentos que dependen de condiciones de vapor atómico bien definidas.
